```python
season_stats = pd.read_csv('./rawdata/Seasons_Stats_edited.csv', skipinitialspace=True)
```

```python
season_stats
```

```python
season_stats['names'] = season_stats['Unnamed: 3']
```

```python
season_stats = season_stats.drop(columns=['Unnamed: 0','Unnamed: 3','Year','Age', 'Tm', 'GS','MP', '3PAr', 'ORB%','DRB%', 'blanl', 'OWS', 'DWS', 'WS/48', 'blank2', 'OBPM', 'DBPM','2P', '2PA', '2P%', 'eFG%','ORB', 'DRB', 'PF'])
```

```python
season_stats
```

```python
season_stats = season_stats.fillna(season_stats.mean())
```

```python
season_stats = season_stats.groupby('names').agg({'G': 'sum', 'PER': 'mean', 'TS%': 'mean', 'FTr':'mean', 'TRB%': 'mean',
                                                   'AST%': 'mean', 'STL%': 'mean', 'BLK%': 'mean', 'TOV%': 'mean', 'USG%':'mean', 
                                                   'WS':'sum' , 'BPM': 'mean', 'VORP':'mean', 'FG' : 'sum', 'FGA': 'sum', 
                                                   'FG%': 'mean', '3P': 'sum', '3PA': 'sum', '3P%' : 'mean', 'FT': 'sum', 'FTA': 'sum',
                                                   'TRB':'sum', 'AST': 'sum', 'STL': 'sum', 'BLK':'sum', 'TOV': 'sum', 'PTS':'sum'})
```

```python
season_stats
```

```python
season_stats['PPG'] = season_stats['PTS'] / season_stats['G']
season_stats['RPG'] = season_stats['TRB'] / season_stats['G']
season_stats['APG'] = season_stats['AST'] / season_stats['G']
season_stats['SPG'] = season_stats['STL'] / season_stats['G']
season_stats['BPG'] = season_stats['BLK'] / season_stats['G']
```

```python
season_stats = season_stats.drop(columns= ['PTS', 'TRB', 'AST','STL', 'BLK'])
```

```python
season_stats
```

```python
season_stats['name'] = season_stats.index
```

```python
season_stats
```

```python
hallOfFamers = pd.merge(hallOfFamers, season_stats, how='left', on='name')
```

```python
hallOfFamers
```
